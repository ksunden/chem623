# Derivation

$\dot{c}_i = -i\omega_{i}c_i - \frac{i}{\hbar}V_{ik}c_k - \Gamma_ic_i$

$\dot{c}_j = -i\omega_{j}c_j - \frac{i}{\hbar}V_{jk}c_k - \Gamma_jc_j$

$\dot{c}_j^* = i\omega_{j}c_j^* + \frac{i}{\hbar}V_{jk}c_k^*- \Gamma_jc_j^*$

$\dot{\rho}_{ij} = \dot{c}_ic_j^* + c_i\dot{c}_j^*$

$\dot{\rho}_{ij} = (-i\omega_{i}c_i - \frac{i}{\hbar}V_{ik}c_k - \Gamma_ic_i)c_j^* + c_i(i\omega_{j}c_j^* + \frac{i}{\hbar}V_{jr}c_k^* - \Gamma_jc_j^*)$

$\dot{\rho}_{ij} = -i\omega_{i}\rho_{ij} - \frac{i}{\hbar}V_{ik}\rho_{kj} - \Gamma_i\rho_{ij} + i\omega_{j}\rho_{ij} + \frac{i}{\hbar}V_{jk}\rho_{ik} - \Gamma_j\rho_{ij}$

Define: $\omega_{ij} \equiv \omega_i - \omega_j$ and $\Gamma_{ij} \equiv \Gamma_i + \Gamma_j$

$\dot{\rho}_{ij} = -i\omega_{ij}\rho_{ij} - \frac{i}{\hbar}(V_{ik}\rho_{kj} - V_{jk}\rho_{ik}) - \Gamma_{ij}\rho_{ij}$

Define: $\delta_{ij} \equiv \omega_{ij} - i\Gamma_{ij}$

$\dot{\rho}_{ij} = -i\delta_{ij}\rho_{ij} + \frac{i}{2}(e^{-i\omega t} + e^{i\omega t})(\Omega_{ij}\rho_{kj} + \Omega_{jk}\rho_{ik})$

Let $\rho_{ij} = \tilde{\rho}_{ij}e^{-i\omega_{RW}t}$

$\dot{\tilde{\rho}}_{ij}e^{-i\omega_{RW}t} - i\omega_{RW}\tilde{\rho}_{ij}e^{-i\omega_{RW}t} = -i\delta_{ij}\tilde{\rho}_{ij}e^{-i\omega_{RW}t} + \frac{i}{2}(e^{-i\omega t} + e^{i\omega t})(\Omega_{ik}\rho_{kj} + \Omega_{jk}\rho_{ik})$

$\dot{\tilde{\rho}}_{ij} = i\omega_{RW}\tilde{\rho}_{ij} - i\delta_{ij}\tilde{\rho}_{ij} + \frac{i}{2}(e^{-i(\omega-\omega_{RW})t} + e^{i(\omega+\omega_{RW}t})(\Omega_{ik}\tilde{\rho}_{kj}e^{-i\omega_{kj}t} + \Omega_{jk}\tilde{\rho}_{ik}e^{-i\omega_{ik}t})$

Neglect the $\omega+\omega_{RW}$ term

$\dot{\tilde{\rho}}_{ij} = i(\omega_{RW}-\delta_{ij})\tilde{\rho}_{ij} + \frac{i}{2}(\Omega_{ik}\tilde{\rho}_{kj}e^{-i(\omega - \omega_{RW} + \omega_{kj})t} + \Omega_{jk}\tilde{\rho}_{ik}e^{-i(\omega - \omega_{RW} + \omega_{ik})t})$

Take the Ket side transition:

$\dot{\tilde{\rho}}_{ij} = i(\omega + \omega_{kj} - \delta_{ij})\tilde{\rho}_{ij} + \frac{i\Omega_{ik}}{2}\tilde{\rho}_{kj}$ and $\rho_{ij} = \tilde{\rho}_{ij} e^{-i(\omega+\omega_{kj})t}$

Assume Steady state:

$\tilde{\rho}_{ij} = -\frac{i\Omega_{ik}}{2i(\omega + \omega_{kj} - \delta_{ij})}\tilde{\rho}_{kj}$

$\tilde{\rho}_{ij} = -\frac{i\Omega_{ik}}{2i(\omega + \omega_{kj} - \omega_{ij} + i\Gamma_{ij})}\tilde{\rho}_{kj} = \frac{\Omega_{ik}}{2(\omega_{ik} - \omega - i\Gamma_{ij})}\tilde{\rho}_{kj}$

For the first equation: 

$\rho_{ij} = \frac{\Omega_{ik}}{2(\omega_{ik} - \omega - i\Gamma_{ij})}e^{i(kz-\omega t)}\rho_{kj}$

$c_ic_j^* = \frac{\Omega_{ik}}{2(\omega_{ik} - \omega - i\Gamma_{ij})}e^{i(kz-\omega t)}c_kc_j^*$

$c_i = \frac{\Omega_{ik}}{2(\omega_{ik} - \omega - i\Gamma_{ij})}e^{i(kz-\omega t)}c_k$

For the second, we say that the transition was being driven at $\omega_1$ and is now being driven at $\omega_2$:

$\tilde{\rho}_{ij} = -\frac{i\Omega_{ik}}{2i(\omega_2 + \omega_{1} - \omega_{ij} + i\Gamma_{ij})}\tilde{\rho}_{kj} = \frac{\Omega_{ik}}{2(\omega_{ij} - \omega_1 - \omega_2 - i\Gamma_{ij})}\tilde{\rho}_{kj}$

$\rho_{ij} = \frac{\Omega_{ik}}{2(\omega_{ij} - \omega_2 - \omega_1 - i\Gamma_{ij})}e^{i((k_1-k_2)z-(\omega_1 + \omega_2)t)}\rho_{kj}$

$c_ic_j^* = \frac{\Omega_{ik}}{2(\omega_{ij} - \omega_2 - \omega_1 - i\Gamma_{ij})}e^{i((k_1-k_2)z-(\omega_1 + \omega_2)t)}c_kc_j^*$

$c_i = \frac{\Omega_{ik}}{2(\omega_{ij} - \omega_2 - \omega_1 - i\Gamma_{ij})}e^{i((k_1-k_2)z-(\omega_1 + \omega_2)t)}c_k$

# Simulation

```python
wavelength = 0.5  # um

# source
n_sources = 2
d_sources = 5  # um

def interfere(n_sources, d_sources, intensities=None, dist=1e5):
    if intensities is None:
        intensities = np.ones(n_sources)
    source_coords = np.zeros((3, n_sources)) # The extra parentheses are needed here because np.zeros expects a shape
                                             # A shape is a tuple (https://docs.python.org/3/tutorial/datastructures.html#tuples-and-sequences)
                                             # A tuple is basically an ordered sequence of python data.
                                             # It is similar to a list, but cannot be modified.
    source_coords[2] = np.arange(n_sources) * d_sources
    # the first dimension of coordinate arrays corresponds to the x, y, and z positions
    # so the source array will have shape (3, n_sources)

    # screen
    dist = 1e5  # um
    n_screen_points = 2000
    theta = 0.0002
    screen_coords = np.zeros((3, n_screen_points))
    screen_coords[0] = dist
    screen_coords[2] = dist*np.tan(np.arange(n_screen_points)*theta)

    # distances between source and screen
    # The T means transpose, it makes sure the coordinates are in the correct orientation
    delta = scipy.spatial.distance_matrix(source_coords.T, screen_coords.T)
    print(delta.shape)
    # the delta array has shape (n_sources, n_screen_points)

    # electric field
    E = np.sum(intensities[:,None] * np.exp(2*np.pi*1j*delta/wavelength), axis=0)
    # the electric field is calculated for each screen point

    # normalized intensity
    I = np.real(E*np.conj(E) / (np.sum(intensities))**2)

    # plot
    plt.plot(screen_coords[2], I)
    plt.xlabel('distance (um)')
    plt.ylabel('intensity')
    plt.grid()
    
```

## Point Sources
### sims

```python
interfere(2, 5)
interfere(2, 20)
interfere(8, 5)
```

### discussion
The distance between the closest points determines the spacing of the fringes: the first and third have fringes in the same locations, the second has fringes much closer together
The distance between the furthest points determines the sharpness of the fringes: the second and third are much sharper than the first, as they have more distance between furthest points.

## Gaussian beam in far field (20cm)

```python
n_sources = 10000
d_sources = .001
coords = np.arange(n_sources) * d_sources
ints = np.sqrt(2 * np.pi) *np.exp(-4 * np.log(2) * ((coords - coords[coords.size // 2]) / 5) ** 2)
interfere(n_sources, d_sources, ints, dist=2e4)
plt.plot(coords, ints/np.max(ints))
plt.xlim(-100,10000)
```

## Lens

```python
xo = 20e4
xi = 10e4
```

```python
beam_waist = 5 #um
wavelength = 0.5 #um
rayleigh_range = np.pi * beam_waist ** 2 / wavelength
print(rayleigh_range, 'µm')
```

```python
focal_length = (rayleigh_range ** 2 + xo * xi)/(xo + xi)

print(focal_length, "µm")
print(focal_length / 1e4, 'cm')
```

```python
mag = 1 / np.sqrt((1 - xo/focal_length) ** 2 + (rayleigh_range / focal_length) ** 2)
print(mag, 'magnification')
print(mag * 5, 'beam waist, µm')
```

```python
waist_lens = beam_waist * (1 + (wavelength * xo / np.pi / beam_waist ** 2) ** 2) ** 0.5
print(waist_lens, 'µm, waist at the lens')

NA = np.sin(np.arctan(waist_lens / (2 * focal_length)))
print(NA, 'NA')
print(focal_length / waist_lens, 'f/ number')
```

The beam waist is smaller than 5 microns, as the lens focuses the beam more tightly. (about 2.5 microns, though a tad more, since it's a gaussian beam, not simple ray optics)
